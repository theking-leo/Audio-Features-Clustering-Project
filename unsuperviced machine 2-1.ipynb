#Exploring my KMeans result

```python
# Compute the first two principal components of the scaled features
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)

# Create a DataFrame with the principal components and cluster labels
pca_df = pd.DataFrame(data=principal_components, columns=[
                      'Principal Component 1', 'Principal Component 2'])
pca_df['Cluster Label'] = clusters

# Create a scatter plot using Seaborn and Matplotlib
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2',
                hue='Cluster Label', data=pca_df, palette='Set1', s=100)
plt.title('KMeans Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
pca = PCA(n_components=3)
principal_components = pca.fit_transform(scaled_df)

# Create a DataFrame with the principal components and cluster labels
pca_df = pd.DataFrame(data=principal_components, columns=[
                      'Principal Component 1', 'Principal Component 2', 'Principal Component 3'])
pca_df['Cluster Label'] = clusters

# 3D Plot using Matplotlib
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot the data points in 3D space
for label in np.unique(clusters):
    ax.scatter(pca_df.loc[clusters == label, 'Principal Component 1'],
               pca_df.loc[clusters == label, 'Principal Component 2'],
               pca_df.loc[clusters == label, 'Principal Component 3'],
               label=f'Cluster {label}', s=100)

ax.set_title('KMeans Clustering in 3D')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.legend(title='Cluster', loc='upper right')
plt.show()


```

```python
# Spider Plot using Seaborn
# Assuming you have a DataFrame called 'scaled_df' with the scaled features

# Calculate the mean values of each feature for each cluster
cluster_means = scaled_df.groupby(clusters).mean()

# Rescale the cluster_means to the original scale
cluster_means_rescaled = scaler.inverse_transform(cluster_means)

# Create a DataFrame for the rescaled cluster_means
cluster_means_df = pd.DataFrame(
    cluster_means_rescaled, columns=columns_selected)

# Normalize the cluster_means_df to a scale of 0 to 1
normalized_cluster_means = (cluster_means_df - cluster_means_df.min()) / \
    (cluster_means_df.max() - cluster_means_df.min())

# Create a spider plot using Seaborn
plt.figure(figsize=(20, 10))
categories = list(normalized_cluster_means)
values = normalized_cluster_means.iloc[clusters].values.tolist()

N = len(categories)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

for cluster in range(len(np.unique(clusters))):
    values_cluster = values[cluster]
    values_cluster += values_cluster[:1]

    ax = plt.subplot(1, len(np.unique(clusters)), cluster + 1, polar=True)
    plt.xticks(angles[:-1], categories, color='grey', size=8)
    plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.50", "0.75"],
               color="grey", size=7)
    plt.ylim(0, 1)
    ax.plot(angles, values_cluster, linewidth=1, linestyle='solid')
    ax.fill(angles, values_cluster, alpha=0.1)

    ax.set_title(f'Cluster {cluster}', size=11, y=1.1)

plt.tight_layout()
plt.show()
```

```python
inertia_values = []
for num_clusters in range(1, 35):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(scaled_df)
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values
plt.figure(figsize=(8, 6))
plt.plot(range(1, 35), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Inertia vs. Number of Clusters')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
# Fit KMeans with n_clusters=20
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(scaled_df)

# Calculate the Silhouette Score
silhouette_avg = silhouette_score(scaled_df, kmeans.labels_)
print("The Silhouette Score is:", silhouette_avg)
```

```python
max_k = 20
sil_score = []

for k in range(2, max_k):
    labels = KMeans(n_clusters=k).fit(scaled_df).labels_
    sil_score.append(silhouette_score(scaled_df, labels))

# Plot the Silhouette Score evolution
plt.title(f'Silhouette Score Evolution from 2 Clusters to {max_k-1} Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.plot(range(2, max_k), sil_score, marker='o')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np
```

```python
max_k = 20
data = scaled_df

for k in range(2, max_k):
    kmean_model = KMeans(n_clusters=k).fit(data)
    labels = kmean_model.labels_

    if 1 < k < data.shape[0]:

        fig, axis1 = plt.subplots(1, 1)
        fig.set_size_inches(8, 6)

        sil_avg = silhouette_score(data, labels)
        print(
            f"* For k = {k} the average silhouette score is: {round(sil_avg, 4)}")
        sample_sil_val = silhouette_samples(data, labels)

        y_lower = 10

        for i in range(k):
            ith_cluster_sv = sample_sil_val[labels == i]
            print(
                f"\t- For cluster {i}, the silhouette value is: {round(np.mean(ith_cluster_sv), 2)}")

            ith_cluster_sv.sort()

            # Calculate where to put the silhouette on the y axes
            ith_cluster_size = ith_cluster_sv.shape[0]
            y_upper = y_lower + ith_cluster_size

            # Paint the cluster
            axis1.fill_betweenx(np.arange(y_lower, y_upper),
                                0, ith_cluster_sv,
                                alpha=0.7)

            # Label the cluster
            axis1.text(-0.05, y_lower + 0.5 * ith_cluster_size, str(i))

            # Calculate the next y lower value for the next cluster of the graph
            y_lower = y_upper + 10  # we leave 10 spaces without any observation

        axis1.set_title("Silhouette score for k = %s" % str(k))
        axis1.set_xlabel("S(i)")
        axis1.set_ylabel("Cluster ID")

        plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score

# Calculate Davies-Bouldin Index for different k values
max_k = 20
davies_bouldin_scores = []
for k in range(2, max_k):
    labels = KMeans(n_clusters=k).fit(scaled_df).labels_
    davies_bouldin_scores.append(davies_bouldin_score(scaled_df, labels))

# Plot the Davies-Bouldin Index scores
plt.title(f'Davies-Bouldin Index score for different cluster sizes')
sns.lineplot(x=range(2, max_k), y=davies_bouldin_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Davies-Bouldin Index')
plt.show()
```

```python

import pandas as pd
import io
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Upload the data file
uploaded = files.upload()

for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))

# Take the first (and only) file from the uploaded files
filename = list(uploaded.keys())[0]

# Convert the binary data to a dataframe
medium_songs_df = pd.read_csv(io.StringIO(uploaded[filename].decode('utf-8')))

# Drop irrelevant columns
medium_songs_df.drop(columns=['mode', 'liveness', 'type',
                              'duration_ms', 'time_signature', 'id', 'html'], inplace=True)

# Define the features for clustering
features = ['danceability', 'energy', 'valence', 'tempo',
            'acousticness', 'key', 'loudness', 'instrumentalness', 'speechiness']

# Perform feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(medium_songs_df[features])

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)  # You can adjust the number of components as needed
pca_features = pca.fit_transform(scaled_features)

# Create a new DataFrame with PCA features
pca_df = pd.DataFrame(data=pca_features, columns=['PCA1', 'PCA2'])

# Perform K-means clustering with 3 clusters (you can change the number of clusters)
num_clusters = 15
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
pca_df['cluster_label'] = kmeans.fit_predict(pca_df)

# Merge the cluster labels back to the original dataframe
medium_songs_df['cluster_label'] = pca_df['cluster_label']

# Print the cluster centers (in the PCA space)
print("Cluster centers (PCA space):")
print(kmeans.cluster_centers_)

# Create playlists for each cluster
for cluster_label in range(num_clusters):
    cluster_songs = medium_songs_df[medium_songs_df['cluster_label']
                                    == cluster_label]
    playlist_name = f"Playlist_{cluster_label + 1}"
    print(f"\n{playlist_name} - Songs:")
    print(cluster_songs['name'].to_list())
```
